The following code: models the two-slit experiment using the following formulae

$$I = I_0[\frac{sin(\alpha)}{\alpha}][{cos(\beta)}]^2$$

$$\alpha = \frac{\pi a}{\lambda} sin\theta$$

$$\beta = \frac{\pi d}{\lambda} sin\theta$$

---

I_0 is the initial intensity

I is the intensity at position y

$a$ = 0.09mm is the width of one of the slits

d = 0.4 mm is the distance between the slits

$\lambda$ = 670 nm is the wavelength of the light

Task 1: plot the predicted 1-D intensity
pattern I/I0 of light observed from two slits.

$$\frac{I}{I_0} = [\frac{sin(\alpha)}{\alpha}][{cos(\beta)}]^2$$

The intensity ratio will be considered here as "I"


```python
#imported as required

import numpy as np
import matplotlib.pyplot as plt

#Set limits for y values (x - axis)
ymin=-.6 
ymax= .6

#Set up an array of angle values between defined y values in mm
angle = np.linspace(np.deg2rad(ymin), np.deg2rad(ymax), 1000)

# Define known values
L = 480                         # length in mm
a = 0.09                        # Slit width in mm
d = 0.4                         # Slit spacing in mm
lam = 670e-6                    # Wavelength in nm

# Calculate values for position y, alpha and beta
y = np.tan(angle)*L
alpha = (np.pi*a/lam)*np.sin(angle)
beta = (np.pi*d/lam)*np.sin(angle)

# Intensity ratio equation
I = ((np.sin(alpha)/alpha)**2)*((np.cos(beta))**2)


# Plot the graph of intensity versus displacement
plt.plot(y, I)

#Add a title and label the axes
plt.title("Graph of Intensity versus Position")
plt.xlabel("y Position in mm")
plt.ylabel("Intensity ratio")
plt.grid()
plt.show()
```

Task 2: On the same figure now plot the predicted 1-D intensity pattern of light
observed from just one of the slits.

This will just be the diffraction component 

$$I_d = I_0[\frac{sin(\alpha)}{\alpha}]$$

```python
#Diffraction Term
I_d = (np.sin(alpha)/alpha)**2

#Plot single and two slits on same graph
plt.plot(y, I, label = "Two slits")
plt.plot(y, I_d, label = "One slit")

#Add a title and label the axes
plt.title("Graph of Intensity versus Position")
plt.xlabel("y Position in mm")
plt.ylabel("Intensity ratio")
plt.legend()
plt.grid()
```

Task 3 simulate the appearance of the pattern in 2-D
on the viewing card in the experiment.

```python
#Create a matrix of values for 1-D pattern repeated in vertical direction
I_matrix = [I for i in range(1000)]

# Create an image of the matrix of values
plt.imshow((I_matrix), aspect=2/3)
plt.title("Simulated Pattern appearance in 2D viewing card")
plt.xlabel("y position mm e-2")
plt.ylabel("z position mm e-2")
```

Task 4: obtain a final 2-D plot that
looks like that observed in the two-slit experiment taking diffraction fall off in z direction into account


Diffraction component:
$$I_d = I_0[\frac{sin(\alpha)}{\alpha}]$$

Altered
$$I_A = I_0[\frac{sin(\sigma)}{\sigma}]$$

sigma term
$$\sigma = \frac{\pi b}{\lambda} sin\theta$$

cos term is interference term
multiplied by sink squared function (a decaying fall off )
want in terms of b
want graph of z versus y

```python
# Alter the Diffraction formula for z direction 

b = 0.06                                   # Adjusted value for b in mm

sigma = (np.pi*b/lam)*np.sin(angle)        # Adjusted alpha term

I_z = (np.sin(sigma)/sigma)**2             # Intensity fall off in z-direction


# Find transpose of fall off and multiply it by I

I_z_matrix = [I_z for i in range(1000)]

I_z_transpose = np.transpose(I_z_matrix)


Light_Observed = (I_z_transpose*(I_matrix))


# Image data

plt.imshow(Light_Observed, aspect=2/3)
plt.title("Actual Pattern appearance in 2D viewing card")
plt.xlabel("y position mm e-2")
plt.ylabel("z position mm e-2")
```
